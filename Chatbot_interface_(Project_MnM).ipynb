```python
from llama_index.llms.huggingface import HuggingFaceLLM
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.core import Settings
import torch

MODEL_NAME = 'TinyLlama/TinyLlama-1.1B-Chat-v1.0'
print(f'Loading {MODEL_NAME}...\n')

llm = HuggingFaceLLM(
    model_name=MODEL_NAME,
    tokenizer_name=MODEL_NAME,
    context_window=2048,
    max_new_tokens=512,
    generate_kwargs={'temperature': 0.3, 'do_sample': True},
    device_map='auto',
    model_kwargs={'torch_dtype': torch.float16}
)

embed_model = HuggingFaceEmbedding(model_name='BAAI/bge-small-en-v1.5')

Settings.llm = llm
Settings.embed_model = embed_model
Settings.chunk_size = 512
Settings.chunk_overlap = 50

print('✅ TinyLlama + BGE embeddings ready.')
if torch.cuda.is_available():
    print(f'   GPU: {torch.cuda.get_device_name(0)}')
```

```python
import gradio as gr
import pdfplumber
import tempfile, os, datetime
from llama_index.core import VectorStoreIndex, Document
from llama_index.core.response_synthesizers import get_response_synthesizer, ResponseMode

# ── Global state ─────────────────────────────────────────────────────────────
current_index = None
loaded_filenames = []


# ── Helper: extract text from PDFs ───────────────────────────────────────────
def load_pdfs(pdf_files):
    """Accept one or multiple PDFs, return list of LlamaIndex Documents."""
    docs = []
    names = []
    for f in pdf_files:
        fname = os.path.basename(f.name)
        names.append(fname)
        with pdfplumber.open(f.name) as pdf:
            for i, page in enumerate(pdf.pages):
                text = (page.extract_text() or '').strip()
                if text:
                    docs.append(Document(
                        text=text,
                        metadata={
                            'source': fname,
                            'page': i + 1
                        }
                    ))
    return docs, names


# ── Helper: temperature by mode ───────────────────────────────────────────────
MODE_TEMPS = {
    'Precise':  0.1,
    'Balanced': 0.3,
    'Creative': 0.7
}

MODE_RESPONSE = {
    'Precise':  ResponseMode.COMPACT,
    'Balanced': ResponseMode.COMPACT,
    'Creative': ResponseMode.TREE_SUMMARIZE
}


# ── Core functions ────────────────────────────────────────────────────────────
def upload_documents(pdf_files, status):
    """Process uploaded PDFs and build the vector index."""
    global current_index, loaded_filenames

    if not pdf_files:
        return '⚠️ No files uploaded.'

    status = '⏳ Processing PDFs and building index...'
    yield status

    try:
        docs, names = load_pdfs(pdf_files)
        loaded_filenames = names

        if not docs:
            yield '❌ No text could be extracted from the PDFs.'
            return

        current_index = VectorStoreIndex.from_documents(docs)
        yield f'✅ Ready — loaded {len(docs)} page(s) from: {', '.join(names)}'

    except Exception as e:
        yield f'❌ Error: {str(e)}'


def chat(user_message, history, mode):
    """Answer a question using the loaded index."""
    global current_index

    if not user_message.strip():
        yield history, ''
        return

    if current_index is None:
        history = history + [[user_message, '⚠️ Please upload a PDF first using the panel on the left.']]
        yield history, ''
        return

    # Show user message immediately
    history = history + [[user_message, '⏳ Thinking...']]
    yield history, ''

    try:
        # Update temperature based on mode
        Settings.llm.generate_kwargs['temperature'] = MODE_TEMPS.get(mode, 0.3)

        retriever = current_index.as_retriever(similarity_top_k=4)
        nodes = retriever.retrieve(user_message)

        synthesizer = get_response_synthesizer(
            response_mode=MODE_RESPONSE.get(mode, ResponseMode.COMPACT)
        )
        response = synthesizer.synthesize(user_message, nodes=nodes)
        answer = response.response.strip()

        # Append source references
        sources = list({f"{n.metadata.get('source','?')} p.{n.metadata.get('page','?')}" for n in nodes})
        source_line = f"\n\n📄 Sources: {', '.join(sources)}"

        history[-1][1] = answer + source_line

    except Exception as e:
        history[-1][1] = f'❌ Error generating answer: {str(e)}'

    yield history, ''


def save_chat_history(history):
    """Save the chat history to a downloadable .txt file."""
    if not history:
        return None

    lines = []
    lines.append(f'DocChat History — {datetime.datetime.now().strftime("%Y-%m-%d %H:%M")}\n')
    lines.append(f'Documents: {', '.join(loaded_filenames) if loaded_filenames else 'None'}\n')
    lines.append('=' * 60 + '\n')

    for i, (q, a) in enumerate(history, 1):
        lines.append(f'Q{i}: {q}')
        lines.append(f'A{i}: {a}')
        lines.append('')

    tmp = tempfile.NamedTemporaryFile(delete=False, suffix='.txt', mode='w')
    tmp.write('\n'.join(lines))
    tmp.close()
    return tmp.name


def clear_chat():
    return [], ''


# ── Gradio UI ─────────────────────────────────────────────────────────────────
with gr.Blocks(
    title='DocChat',
    theme=gr.themes.Soft(primary_hue='violet', neutral_hue='slate'),
    css="""
        .chatbot { font-size: 14px; }
        .status-box { font-size: 13px; }
        footer { display: none !important; }
    """
) as demo:

    gr.Markdown("""
    # 🤖 DocChat
    **Ask questions about your PDF documents — powered by TinyLlama + RAG**
    """)

    with gr.Row():

        # ── Left panel — upload & settings ───────────────────────────────────
        with gr.Column(scale=1, min_width=260):
            gr.Markdown('### 📁 Documents')

            pdf_upload = gr.File(
                label='Upload PDFs (one or more)',
                file_types=['.pdf'],
                file_count='multiple'
            )

            upload_btn = gr.Button('📥 Load Documents', variant='primary')

            upload_status = gr.Textbox(
                label='Status',
                value='No documents loaded yet.',
                interactive=False,
                elem_classes=['status-box']
            )

            gr.Markdown('### ⚙️ Settings')

            mode_dropdown = gr.Dropdown(
                choices=['Precise', 'Balanced', 'Creative'],
                value='Balanced',
                label='Response mode',
                info='Precise = factual & short | Creative = longer & exploratory'
            )

            gr.Markdown('### 💾 Export')

            save_btn = gr.Button('⬇️ Download chat history')
            download_file = gr.File(label='Download', visible=False)

        # ── Right panel — chat ────────────────────────────────────────────────
        with gr.Column(scale=3):
            chatbot = gr.Chatbot(
                label='DocChat',
                height=460,
                elem_classes=['chatbot'],
                bubble_full_width=False,
                show_copy_button=True
            )

            with gr.Row():
                msg_input = gr.Textbox(
                    placeholder='Ask a question about your document...',
                    label='',
                    scale=5,
                    lines=1
                )
                send_btn = gr.Button('Send', variant='primary', scale=1)

            clear_btn = gr.Button('🗑️ Clear chat', size='sm')

            gr.Examples(
                examples=[
                    ['What is this document about?'],
                    ['Summarize the key specifications.'],
                    ['What are the storage conditions?'],
                    ['List all part or lot numbers mentioned.'],
                    ['What sterilization method was used?']
                ],
                inputs=msg_input,
                label='Example questions'
            )

    # ── Event wiring ──────────────────────────────────────────────────────────
    upload_btn.click(
        fn=upload_documents,
        inputs=[pdf_upload, upload_status],
        outputs=upload_status
    )

    send_btn.click(
        fn=chat,
        inputs=[msg_input, chatbot, mode_dropdown],
        outputs=[chatbot, msg_input]
    )

    msg_input.submit(
        fn=chat,
        inputs=[msg_input, chatbot, mode_dropdown],
        outputs=[chatbot, msg_input]
    )

    clear_btn.click(
        fn=clear_chat,
        outputs=[chatbot, msg_input]
    )

    def handle_save(history):
        path = save_chat_history(history)
        if path:
            return gr.File(value=path, visible=True)
        return gr.File(visible=False)

    save_btn.click(
        fn=handle_save,
        inputs=chatbot,
        outputs=download_file
    )

demo.launch(share=True, debug=False)
```
