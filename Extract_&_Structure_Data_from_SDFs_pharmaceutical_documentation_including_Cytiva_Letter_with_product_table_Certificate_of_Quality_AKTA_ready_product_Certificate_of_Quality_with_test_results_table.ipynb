```python
!pip install pymupdf
import fitz
doc = fitz.open("sample-sdf-document (1)pharma_project.pdf")
```

```python
all_words = []
for page_num in range(len(doc)):
    page = doc.load_page(page_num)
    words = page.get_text("words")  # Extract words with their bounding boxes
    all_words.extend(words)

# Convert to a list of dictionaries for structured output
structured_words = []
for word_info in all_words:
    # word_info format: (x0, y0, x1, y1, word, block_no, line_no, word_no)
    structured_words.append({
        'text': word_info[4],
        'bbox': word_info[:4]  # (x0, y0, x1, y1)
    })

print(f"Extracted {len(structured_words)} structured words from the document.")
print("First 10 structured words:")
for i in range(min(10, len(structured_words))):
    print(structured_words[i])
```
