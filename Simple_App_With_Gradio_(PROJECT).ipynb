```python
# Cell 2 — Build and launch the app
import gradio as gr
import math

def calculate(number):
    n = int(number)

    if n < 0:
        return "Error: number must be 0 or greater", "", ""

    square    = n * n
    cube      = n * n * n
    factorial = math.factorial(n)

    return square, cube, factorial


app = gr.Interface(
    fn=calculate,
    inputs=gr.Number(label="Enter a number", precision=0, minimum=0),
    outputs=[
        gr.Number(label="Square (n × n)"),
        gr.Number(label="Cube (n × n × n)"),
        gr.Number(label="Factorial (1 × 2 × ... × n)"),
    ],
    title="Math Calculator",
    description="Enter a whole number to see its square, cube, and factorial.",
    examples=[[5], [7], [10]],
)

app.launch(share=True)   # share=True gives you a public link for your phone
```
