```python
import fitz # PyMuPDF
import re
import json # Added to handle JSON output

# Initialize global lists to store extracted data
date_fields = []
table_data_fields = []
vendor_info_fields = []

# Path to your PDF document
pdf_path = '/content/sample-sdf-document (1)pharma_project.pdf'

try:
    doc = fitz.open(pdf_path)
    print(f"Successfully loaded {doc.page_count} pages from {pdf_path}")
except Exception as e:
    print(f"Error loading PDF: {e}")
    doc = None
```

### Step 3: Extract Date Fields

To extract date fields (like Effective Date, Manufacturing Date, Revision Date), we'll iterate through each page and search for common date patterns. We'll use regular expressions to identify potential dates and then check if they are associated with keywords like 'Date', 'Effective', 'Manufacturing', 'Revision', etc. For each found date, we'll extract the text and its bounding box.

```python
if doc:
    global date_fields # Declare as global to modify the list initialized in 44251e7b
    date_fields = [] # Clear previous data if re-running
    # Common date patterns (e.g., MM/DD/YYYY, DD-Mon-YYYY, YYYY-MM-DD)
    # This regex is a general attempt, and might need refinement based on actual date formats in the document
    date_pattern = re.compile(r'\b(?:(?:\d{1,2}[-/]\d{1,2}[-/]\d{2,4})|(?:\d{4}[-/]\d{1,2}[-/]\d{1,2})|(?:\d{1,2} (?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec) \d{2,4}))\b', re.IGNORECASE)
    keyword_pattern = re.compile(r'(effective date|manufacturing date|revision date|date:|release date|expiry date)', re.IGNORECASE)

    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text_blocks = page.get_text("dict")["blocks"]

        for b in text_blocks:
            if b["type"] == 0:  # text block
                for l in b["lines"]:
                    for s in l["spans"]:
                        text = s["text"]
                        bbox = s["bbox"]

                        # Search for keywords and dates in the text
                        keyword_matches = list(keyword_pattern.finditer(text))
                        date_matches = list(date_pattern.finditer(text))

                        if keyword_matches or date_matches:
                            extracted_text = text.strip()
                            date_fields.append({
                                'page': page_num + 1,
                                'text': extracted_text,
                                'bbox': list(bbox) # Convert bbox tuple to list
                            })

    print("\n--- Extracted Date Fields ---")
    if date_fields:
        for item in date_fields:
            print(f"Page {item['page']}: Text='{item['text']}', BBox={item['bbox']}")
    else:
        print("No specific date fields found using the defined patterns and keywords.")
```

### Step 4: Extract Table-based Fields (Test Results/Batch Data)

Extracting table data is more involved as it requires identifying table structures. PyMuPDF can extract text blocks with their bounding boxes, which we can then use to infer table-like structures. For this example, I will demonstrate a general approach of extracting text from a certain rectangular area, which you would define based on where a table is expected.

*Note: A more robust table extraction would involve using dedicated table parsing libraries or more sophisticated layout analysis.*

Let's assume there's a table on a specific page within a given rectangle. We will extract all text within that rectangle.

```python
if doc:
    global table_data_fields # Declare as global to modify the list initialized in 44251e7b
    table_data_fields = [] # Clear previous data if re-running
    # Define a bounding box for a potential table area on a specific page (e.g., page 1)
    # You'll need to adjust these coordinates based on your document's actual layout.
    # For example, rect = fitz.Rect(x0, y0, x1, y1)
    # These are illustrative coordinates; replace them with actual table coordinates.

    # Let's try to extract text from a hypothetical table region on the first page.
    target_page_num = 0  # Page 1
    # Example: a rectangle from (50, 100) to (550, 400) on the page
    table_rect = fitz.Rect(50, 100, 550, 400)

    if target_page_num < len(doc):
        page = doc.load_page(target_page_num)
        # Extract text within the defined rectangle
        text_in_table_area = page.get_text(clip=table_rect)

        # Get text blocks with their bounding boxes that fall within the table_rect
        blocks = page.get_text("dict", clip=table_rect)["blocks"]

        for b in blocks:
            if b["type"] == 0:  # text block
                for l in b["lines"]:
                    for s in l["spans"]:
                        text = s["text"].strip()
                        bbox = s["bbox"]
                        if text:
                            table_data_fields.append({
                                'page': target_page_num + 1,
                                'text': text,
                                'bbox': list(bbox) # Convert bbox tuple to list
                            })

    print("\n--- Extracted Table-based Fields (Hypothetical Area) ---")
    if table_data_fields:
        for item in table_data_fields:
            print(f"Page {item['page']}: Text='{item['text']}', BBox={item['bbox']}")
    else:
        print(f"No significant text found in the hypothetical table area on page {target_page_num + 1} or document not loaded.")
```

### Step 5: Extract Vendor Information from Document Headers

Vendor information is often found in the document headers or footers. We will iterate through the first few pages and search for common keywords like 'Vendor', 'Manufacturer', 'Company Name', or typical address patterns. We will extract the text and its bounding box.

```python
if doc:
    global vendor_info_fields # Declare as global to modify the list initialized in 44251e7b
    vendor_info_fields = [] # Clear previous data if re-running
    # Keywords to look for vendor information
    vendor_keywords = re.compile(r'(vendor|manufacturer|company name|produced by|distributed by|firm name|address)', re.IGNORECASE)
    # Check the first few pages for vendor info
    pages_to_check = min(len(doc), 3)

    for page_num in range(pages_to_check):
        page = doc.load_page(page_num)
        # A common approach is to check the top margin for header info
        # Let's define a header rectangle (e.g., top 100 units of the page)
        header_rect = fitz.Rect(0, 0, page.rect.width, 100)

        text_blocks = page.get_text("dict", clip=header_rect)["blocks"]

        for b in text_blocks:
            if b["type"] == 0: # text block
                for l in b["lines"]:
                    for s in l["spans"]:
                        text = s["text"]
                        bbox = s["bbox"]
                        if vendor_keywords.search(text):
                            vendor_info_fields.append({
                                'page': page_num + 1,
                                'text': text.strip(),
                                'bbox': list(bbox) # Convert bbox tuple to list
                            })

    print("\n--- Extracted Vendor Information ---")
    if vendor_info_fields:
        # Remove duplicates for printing, but keep all in the global list for consolidation
        unique_vendor_info_for_print = []
        seen_texts = set()
        for item in vendor_info_fields:
            if item['text'] not in seen_texts:
                unique_vendor_info_for_print.append(item)
                seen_texts.add(item['text'])

        for item in unique_vendor_info_for_print:
            print(f"Page {item['page']}: Text='{item['text']}', BBox={item['bbox']}")
    else:
        print("No vendor information found in the header areas of the first few pages using the defined keywords.")

    # The doc.close() is handled in the final consolidation cell.
```

### Step 6: Consolidate and Output Results in JSON Format

Now, let's consolidate all the extracted data (date fields, table data, and vendor information) into a single structured JSON output. We'll also ensure the PDF document is closed after all processing is complete.

```python
final_extracted_data = {
    "date_fields": date_fields,
    "table_data_fields": table_data_fields,
    "vendor_info_fields": vendor_info_fields
}

print("\n--- Consolidated JSON Output ---")
print(json.dumps(final_extracted_data, indent=2))

# Close the document after all processing is complete
if doc:
    doc.close()
    print("\nPDF document closed.")
```
